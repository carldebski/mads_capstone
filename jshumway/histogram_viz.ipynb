```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('all_merged_data.csv').drop(columns = ['date'])
data
```

```python
ranges = data.max() - data.min()
```

```python
ranges
```

```python

mean_values = data.mean()

# Plotting the overlaying histograms for ranges and means
plt.figure(figsize=(12, 6)) 

# Plot the histogram for ranges
sns.histplot(ranges, kde=True, color='skyblue', label='Ranges', stat="density", bins=100, alpha=0.5)

# Plot the histogram for means
sns.histplot(mean_values, kde=True, color='green', label='Means', stat="density", bins=100, alpha=0.5)

#title and labels
plt.title("Overlaying Histograms of Ranges and Means for Keywords", fontsize=16)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Density", fontsize=12)

plt.legend()

plt.show()
```
